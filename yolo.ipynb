```python
# load pre-trained yolov11 small model
model = YOLO('yolo11s.pt')
```

```python
# train model on trash dataset
model.train(
    data='taco.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='taco_model'
)
```
